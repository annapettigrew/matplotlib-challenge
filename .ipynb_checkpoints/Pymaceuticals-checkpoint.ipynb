# Pymaceuticals

```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
# Adding CSVs
mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"
```

```python
# Reading CSVs
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
# Combining the CSV's
pharm_df = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])

# Display
pharm_df
```

```python
# Displaying the total number of mice in the dataset
total_mice = len(pharm_df["Mouse ID"])
total_mice
```

```python
# Display the number of unique mice IDs in the data
unique_mice = len(pharm_df["Mouse ID"].unique())
unique_mice
```

```python
# Displaying the number of Mice with Duplicate Timepoints 
pharm_df_duplicates = pharm_df[pharm_df.duplicated(["Mouse ID", "Timepoint"])]
pharm_df_duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID
pharm_df_duplicates_two = pharm_df.set_index('Mouse ID')
pharm_df_duplicates_two.loc['g989']
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
new_pharm_df = pharm_df.drop_duplicates(["Mouse ID", "Timepoint"])
new_pharm_df
```

```python
# Checking the number of mice in the clean DataFrame
new_mice_count = len(new_pharm_df["Mouse ID"])
new_mice_count
```

# Summary Statistics 

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_df = new_pharm_df.groupby(['Drug Regimen'])
tumor_mean = summary_df['Tumor Volume (mm3)'].mean()
tumor_median = summary_df['Tumor Volume (mm3)'].median()
tumor_variance = summary_df['Tumor Volume (mm3)'].var()
tumor_sd = summary_df['Tumor Volume (mm3)'].std()
tumor_SEM = summary_df['Tumor Volume (mm3)'].sem()

# Assemble the resulting series into a single summary DataFrame
summary_stat_df = pd.DataFrame({"Mean":tumor_mean,
                               "Median":tumor_median,
                               "Variance":tumor_variance,
                               "Standard Deviation":tumor_sd,
                               "SEM":tumor_SEM})
summary_stat_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

summary_agg =new_pharm_df.groupby (['Drug Regimen'])['Tumor Volume (mm3)'].agg(['mean','median','var','std','sem'])

summary_agg_df = pd.DataFrame({"Mean":tumor_mean,
                               "Median":tumor_median,
                               "Variance":tumor_variance,
                               "Standard Deviation":tumor_sd,
                               "SEM":tumor_SEM})
summary_agg_df
```

# Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
great_timepoint = new_pharm_df.groupby(['Mouse ID'])['Timepoint'].last().sort_values()
great_timepoint_df = pd.DataFrame(great_timepoint) 

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
new_great_timepoint = pd.merge(new_pharm_df, great_timepoint_df, on='Mouse ID')
new_great_timepoint
```

```python
 # Put treatments into a list for for loop (and later for plot labels)
capomulin = new_pharm_df.loc[new_pharm_df['Drug Regimen'] == 'Capomulin']
ramicane = new_pharm_df.loc[new_pharm_df['Drug Regimen'] == 'Ramicane']
infubinol = new_pharm_df.loc[new_pharm_df['Drug Regimen'] == 'Infubinol']
ceftamin = new_pharm_df.loc[new_pharm_df['Drug Regimen'] == 'Ceftamin']

# list
regimens_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_data = []
outliers = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for regimens in regimens_list:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_mice = new_pharm_df.loc[(new_pharm_df['Drug Regimen'] == regimens)]
    
    # add subset 
    tumor_data.append(new_pharm_df['Tumor Volume (mm3)'])
    outliers.append(new_pharm_df['Tumor Volume (mm3)'])
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_mice['Tumor Volume (mm3)'].quantile([0.25,0.5,0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    
    upper_bound = upperq + (1.5*iqr)
    lower_bound = lowerq - (1.5*iqr)
    
    print(f' {regimens}')
    print(f' -------------')
    print(f' The possible Outliers for {regimens} are: {upper_bound}')
    print(f' The possible Outliers for {regimens} are: {lower_bound}')
    print(f' The IQR for {regimens} is: {iqr}')
    print(f' The Upper Quartile for {regimens} is: {upperq}')
    print(f' The Lower Quartile for {regimens} is: {lowerq}')
    print(f' ------------------------------------------------')
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
x_cap = capomulin['Tumor Volume (mm3)']
x_ram = ramicane['Tumor Volume (mm3)']
x_inf = infubinol['Tumor Volume (mm3)']
x_cef = ceftamin['Tumor Volume (mm3)']
```

# Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
# linear regression
from scipy.stats import linregress
x_value = capomulin_avg["Weight (g)"]
y_value = capomulin_avg["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_value, y_value)
regression_value = x_value * slope + intercept
line_eq = 'y = ' + str(round(slope,2)) + 'x +' + str(round(intercept,2))
plt.scatter(x_value, y_value, color='gold')
plt.title("Average Tumor Volume VS Mouse Weight for Capomulin Regimen")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.grid()
plt.plot(x_value, regression_value, 'sienna')
plt.annotate(line_eq,(16,45), fontsize=14, color='black')
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model 
# Correlation Coefficient
correlation_coe = round(st.pearsonr(capomulin_avg['Weight (g)'], capomulin_avg['Tumor Volume (mm3)'])[0],2)
print(f'The Correlation Coefficient between Mouse Weight and Average Tumor Volume for the Capomulin Regimen is: {correlation_coe}')
```
